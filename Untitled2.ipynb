```python
# ============================================================
#         Q5 - Python is a High-Level Language
# ============================================================

print("--- Q5: What is a High-Level Language? ---")
print("""A HIGH-LEVEL language means the language is closer to human language
(English-like) and far from machine language (0s and 1s).

Key points:
  - Easy to read and write (print("Hello") vs binary instructions)
  - Hides complex hardware details like memory management, CPU registers
  - Portable: same code runs on different machines
  - Python, Java, C++ are high-level; Assembly is low-level
""")
```

```python
# ============================================================
#      Q6 - Compiled vs Interpreted Languages
# ============================================================

print("--- Q6: Compiled vs Interpreted Languages ---")
print("""COMPILED LANGUAGE:
  - Source code is translated ENTIRELY into machine code BEFORE running.
  - A separate executable file is produced.
  - Faster at runtime.
  - Examples: C, C++, Go, Rust

INTERPRETED LANGUAGE:
  - Source code is translated and executed LINE BY LINE at runtime.
  - No separate executable; needs an interpreter every time.
  - Easier to debug but generally slower.
  - Examples: Python, JavaScript, Ruby, PHP

Note: Python is technically both — it compiles to bytecode (.pyc)
first, then the Python interpreter (CPython) executes that bytecode.
""")
```

```python
# ============================================================
#       Q7 - Get RAM Address of All Variables (id())
# ============================================================

print("--- Q7: RAM Addresses using id() ---")

a = 500
b = 400
c = 400
d = 300
name1 = "Mahesh"
name2 = "Mithilesh"
name3 = "Ramesh"
name4 = "Sumesh"

print(f"a    = {a!r:12}  → id: {id(a)}")
print(f"b    = {b!r:12}  → id: {id(b)}")
print(f"c    = {c!r:12}  → id: {id(c)}")
print(f"d    = {d!r:12}  → id: {id(d)}")
print(f"name1= {name1!r:12}  → id: {id(name1)}")
print(f"name2= {name2!r:12}  → id: {id(name2)}")
print(f"name3= {name3!r:12}  → id: {id(name3)}")
print(f"name4= {name4!r:12}  → id: {id(name4)}")

print("\n🔍 Checking for shared addresses (same id = same object in RAM):")
variables = {"a": a, "b": b, "c": c, "d": d,
             "name1": name1, "name2": name2, "name3": name3, "name4": name4}

found = False
names = list(variables.keys())
for i in range(len(names)):
    for j in range(i + 1, len(names)):
        if id(variables[names[i]]) == id(variables[names[j]]):
            print(f"  ✅ '{names[i]}' and '{names[j]}' share the same RAM address! "
                  f"(value={variables[names[i]]!r}, id={id(variables[names[i]])})")
            found = True

if not found:
    print("  No shared addresses found.")

print("\n💡 Why do b and c (both = 400) share the same address?")
print("   Python CACHES small integers (-5 to 256). 400 is outside this range,")
print("   so it may or may not be shared depending on the Python session.")
```
